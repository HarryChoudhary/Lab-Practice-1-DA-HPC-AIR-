```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('diabetes.csv')
df
```

```python
df2=df.drop(['Pregnancies','Outcome'],axis=1)
```

```python
avg_insulin=df['Insulin'].astype('float').mean(axis=0)
```

```python
#find a mean to replace missing values by mean (astype is use to change column type)
avg_insulin=df['Insulin'].astype("float").mean(axis=0)
avg_glucose=df['Glucose'].astype("float").mean(axis=0)
avg_bp=df['BloodPressure'].astype("float").mean(axis=0)
avg_skinthickness=df['SkinThickness'].astype("float").mean(axis=0)
avg_BMI=df['BMI'].astype("float").mean(axis=0)
avg_DPF=df['DiabetesPedigreeFunction'].astype("float").mean(axis=0)
avg_age=df['Age'].astype("float").mean(axis=0)
```

```python
df['Insulin'].replace(0,avg_insulin,inplace=True)
```

```python
#replace zero insulin with mean
#print(avg_2)
df['Insulin'].replace(0,avg_insulin,inplace=True) #make it work to replace zero
df['Glucose'].replace(0,avg_glucose,inplace=True)
df['BloodPressure'].replace(0,avg_bp,inplace=True)
df['SkinThickness'].replace(0,avg_skinthickness,inplace=True)
df['BMI'].replace(0,avg_BMI,inplace=True)
df['DiabetesPedigreeFunction'].replace(0,avg_DPF,inplace=True)
df['Age'].replace(0,avg_age,inplace=True)
```

```python
print(df.head())
```

```python
x=df.drop('Outcome',axis=1)
```

```python
y=df['Outcome']
```

```python
print(y.head())
```

```python
scaler=MinMaxScaler()
```

```python
scal_x=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(scal_x,y,test_size=0.2,random_state=20)
```

```python
classifier=GaussianNB()
```

```python
classifier.fit(x_train,y_train)
```

```python
y_predict=classifier.predict(x_test)
```

```python
c=confusion_matrix(y_test,y_predict)
```

```python
print(c)
```

```python
cf=pd.DataFrame(c,columns=['No','yes'],index=['actual no','actual yes'])
```

```python
score=precision_recall_fscore_support(y_test,y_predict,average='weighted')
```
